```python
object_names = set("/c/en/%s" % "_".join(s.split()) for s in objects) | set("/c/en/%s/n" % "_".join(s.split()) for s in objects)
len(object_names)
```

```python
df = pd.read_table("../data/conceptnet-assertions-5.6.0.csv.gz", compression='gzip', chunksize=1000000, names=['url', 'relation', 'start', 'end', 'data'])

tuples = []

for chunk in df:    
    for row in tqdm.tqdm(chunk.itertuples(), total=1000000):
        start = row.start
        end = row.end
        rel = row.relation
        
        if start in object_names and end in object_names:
            print(start, rel, end)
            tuples.append((start, rel, end))
```

```python
df = pd.read_table("../data/conceptnet-assertions-5.6.0.csv.gz", compression='gzip', chunksize=1000000, names=['url', 'relation', 'start', 'end', 'data'])

tuples = []

for chunk in df:    
    for row in tqdm.tqdm(chunk.itertuples(), total=1000000):
        start = row.start
        end = row.end
        rel = row.relation
        
        if start in object_names and end not in object_names and rel == '/r/IsA' and end.startswith('/c/en'):
            print(start, rel, end)
            object_names.add(end)
```

```python
tuples = json.load(open("../data/cnetrel.json"))
```

```python
import collections

conceptnet = collections.defaultdict(lambda: set())

for start, rel, end in tuples:
    start = start.split("/")[3]
    end = end.split("/")[3]
    rel = rel.split("/")[-1]
    
    if start != end:
        conceptnet[rel].add((start, end))
```

```python
conceptnet = dict(conceptnet)
len(conceptnet)
```

```python
with open('../data/conceptnet.json', "w") as fp:
    json.dump({k: list(v) for k,v in conceptnet.items()}, fp)
```

```python
with open("../data/conceptnet.json") as fp:
    conceptnet = json.load(fp)
```

```python
import networkx
```

```python
G = networkx.DiGraph()

is_a  = conceptnet['IsA']

for src,dst in is_a:
    G.add_edge(src,dst)
```

```python
_, scores = networkx.hits(G)
# scores = networkx.pagerank(G)
best_nodes = set(list(sorted(scores, key=scores.get, reverse=True))[:100])
```

```python
G = networkx.DiGraph()
```

```python
for src,dst in is_a:
    if src not in best_nodes or dst not in best_nodes:
        continue
        
    if scores[src] < scores[dst]:
        G.add_edge(src,dst)
```

```python
dot = networkx.nx_pydot.to_pydot(G)
dot.set_rankdir("LR")
```

```python
dot.write_pdf('../data/graph.pdf')
```
