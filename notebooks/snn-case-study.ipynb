# Usando SNNs para problemas de Transfer Learning

En este ejemplo usaremos una SNN entrenada en un modelo particular (**a elegir todavía**) para luego hacer _fine-tunning_ en un problema de transfer learning distinto. 

```python
from snn import Entity, Relation
from snn.models import SNNModel
from keras.layers import Input, Dense
from keras.models import Model
from snn.utils import draw
```

```python
A = Entity("A")
B = Entity("B")

r1 = Relation("r1", A, B)
r2 = Relation("r2", A, B)
```

`SNNModel`, a diferencia de la clase base `SNN`, ya engloba todo el proceso de construir la SNN a partir de una entrada predeterminada, y compilar el modelo de entrenamiento.

```python
snn = SNNModel(entities=[A,B], relations=[r1,r2], instances=100)
```

`snn.model` es el modelo de `keras` que se debe usar para pre-entrenar la snn. Esto solo es relevante si se quiere entrenar la red en la base de conocimientos. Si ya está entrenada, se debe usar un método `snn.load(...)` **no implementado todavía** para cargar los pesos pre-entrenados.

```python
draw(snn.model)
```
